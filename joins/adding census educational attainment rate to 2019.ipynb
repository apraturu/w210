```python
df_2019 = pd.read_csv("../final_data/2019_new-var-PUMA_mapped_processed.csv")
df_2019.head()
```

```python
cols = {"Percent-Population 25 years and over-Less than 9th grade": "less_than_9th",
        "Percent-Population 25 years and over-9th to 12th grade, no diploma": "9th_to_12th",
        "Percent-Population 25 years and over-High school graduate (includes equivalency)": "hs_grad",
        "Percent-Population 25 years and over-Some college, no degree": "some_college",
        "Percent-Population 25 years and over-Associate's degree": "associates_degree",
        "Percent-Population 25 years and over-Bachelor's degree": "bach_degree",
        "Percent-Population 25 years and over-Graduate or professional degree": "grad_or_prof_degree",
        "Percent-Population 25 years and over-High school graduate or higher": "hs_or_higher",
        "Percent-Population 25 years and over-Bachelor's degree or higher": "bach_degree_or_higher",
       }
df_hv = pd.read_csv("educational_attainment.csv").drop(columns=['Geographic Area Name']).rename(columns=cols)
df_hv.head()
```

```python
# make a copy of df_2018 that we will join to

df_2019_hv = df_2019.copy()
df_2019_hv.head()
```

```python
# try a join

df_2019_hv = df_2019_hv.join(df_hv.set_index('PUMA'), on='full PUMA')
df_2019_hv.head()
```

```python
# create new csv of processed data
# this creates a csv that is ~150 MB - too big for Git
df_2019_hv.to_csv('../final_data/2019_new-var_educational_attainment.csv',index=False)
```
