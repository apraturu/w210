```python
df_2019 = pd.read_csv("../final_data/2019_new-var_mapped_processed.csv")
df_2019.head()
```

```python
df_inc = pd.read_csv("incarceration_rates_2010.csv")
df_inc.head()
```

```python
# make a copy of df_2018 that we will join to

df_2019_inc = df_2019.copy()
df_2019_inc.head()
```
