```python
# create dataframes for the new variables

# home value
df_hv = pd.read_csv("/Users/alexandrawest/Desktop/W210-Capstone/Census_data/median_home_value_2.csv").drop(columns=['Geographic Area Name', 'Margin of Error!!Median value (dollars)']).rename(columns={"Estimate!!Median value (dollars)": "med_home_value"})

# gross rent
df_gr = pd.read_csv("/Users/alexandrawest/Desktop/W210-Capstone/Census_data/median_gross_rent_2.csv").drop(columns=['Geographic Area Name', 'Margin of Error!!Median gross rent']).rename(columns={"Estimate!!Median gross rent": "med_gross_rent"})

# incarceration rates 
df_inc = pd.read_csv("incarceration_rates_2010.csv")

# unemployment rate
df_unem = pd.read_csv("/Users/alexandrawest/Desktop/W210-Capstone/Census_data/unemployment_2.csv").drop(columns=['Geographic Area Name']).rename(columns={"Percent!!EMPLOYMENT STATUS!!Population 16 years and over!!In labor force!!Civilian labor force!!Unemployed": "unemployment_rate"})

# health insurance coverage
df_hic = pd.read_csv("/Users/alexandrawest/Desktop/W210-Capstone/Census_data/health_insurance_coverage_2.csv").drop(columns=['Geographic Area Name']).rename(columns={"Estimate!!Percent Insured!!Civilian noninstitutionalized population": "insured_rate", "Estimate!!Percent Uninsured!!Civilian noninstitutionalized population": "uninsured_rate"})


# educational attainment
cols = {"Percent-Population 25 years and over-Less than 9th grade": "less_than_9th",
        "Percent-Population 25 years and over-9th to 12th grade, no diploma": "9th_to_12th",
        "Percent-Population 25 years and over-High school graduate (includes equivalency)": "hs_grad",
        "Percent-Population 25 years and over-Some college, no degree": "some_college",
        "Percent-Population 25 years and over-Associate's degree": "associates_degree",
        "Percent-Population 25 years and over-Bachelor's degree": "bach_degree",
        "Percent-Population 25 years and over-Graduate or professional degree": "grad_or_prof_degree",
        "Percent-Population 25 years and over-High school graduate or higher": "hs_or_higher",
        "Percent-Population 25 years and over-Bachelor's degree or higher": "bach_degree_or_higher",
       }
df_edu = pd.read_csv("/Users/alexandrawest/Desktop/W210-Capstone/Census_data/educational_attainment_2.csv").drop(columns=['Geographic Area Name']).rename(columns=cols)
```

```python
# check dataframes

# df_hv.head()
# df_gr.head()
# df_inc.head()
# df_unem.head()
# df_hic.head()
# df_edu.head()
```

```python
# home values is an object - need to convert to float
# the following doesn't work because we have a value: '2,000,000+' which is a string
# df_hv["med_home_value"] = df_hv.med_home_value.astype(float)

df_hv['med_home_value'] = df_hv['med_home_value'].replace(['2,000,000+','2,000,000'],2000000)
df_hv.dtypes
```

```python
#now we can change to float
df_hv['med_home_value'] = df_hv.med_home_value.astype(float)
df_hv.dtypes
```

```python
# make a copy of df_2019 that we will join to

df_2019_all = df_2019.copy()
df_2019_all.head() # should be the same as above
```

```python
# join one at a time, using the df_2019_all dataframe as the final

# join home value
df_2019_all = df_2019_all.join(df_hv.set_index('PUMA'), on='full PUMA')

# adds 1 column, med_home_value
```

```python
# join gross rent
df_2019_all = df_2019_all.join(df_gr.set_index('PUMA'), on='full PUMA')

# adds 1 column, med_gross_rent
```

```python
# join unemployment
df_2019_all = df_2019_all.join(df_unem.set_index('PUMA'), on='full PUMA')

# adds 1 column, unemployment
```

```python
# join health insurance coverage
df_2019_all = df_2019_all.join(df_hic.set_index('PUMA'), on='full PUMA')

# adds 2 column, insured_rate and uninsured_rate
```

```python
# join educational attainment
df_2019_all = df_2019_all.join(df_edu.set_index('PUMA'), on='full PUMA')

# adds 9 columns
```

```python
# join incarceration
df_2019_all = df_2019_all.join(df_inc.set_index('State'), on='state')

# adds 10 columns
```
