# Joining the 2018 data (with new variables) to Zillow home values by state


```python
import pandas as pd
import numpy as np
```

```python
!pwd
```

```python
df_2018 = pd.read_csv("../Census_data_2018/2018_new-var_mapped_processed.csv")
df_2018.head()
```

```python
df_zill = pd.read_csv("Zillow/Metro_zhvi_uc_sfrcondo_tier_0.33_0.67_sm_sa_mon.csv")
df_zill.head()
```

```python
df_zill = df_zill[['RegionID', 'SizeRank', 'RegionName', 'RegionType', 'StateName', '2019-01-31', '2019-02-28', '2019-03-31', '2019-04-30', '2019-05-31', '2019-06-30', '2019-07-31', '2019-08-31', '2019-09-30', '2019-10-31', '2019-11-30', '2019-12-31']]

col = df_zill.loc[: , '2019-01-31':'2019-12-31']

df_zill['home_value'] = col.mean(axis=1)

df_zill_less = df_zill[['StateName', 'home_value']]
```

```python
df_zill_less.head()
```

```python
# make a copy of df_2018 that we will join to

df_2018_zill = df_2018.copy()
df_2018_zill.head()
```

```python
# try a join

df_2018_zill = df_2018_zill.join(df_zill_less.set_index('StateName'), on='state')
df_2018_zill.head()
```

```python
# add rent index

df_zill_rent = pd.read_csv("Zillow/Metro_ZORI_AllHomesPlusMultifamily_SSA.csv")
df_zill_rent.head()
```

```python
df_zill_j = df_zill[['RegionID', 'StateName']].drop_duplicates()
df_zill_j
```

```python
df_zill_rent = df_zill_rent[['RegionID', '2019-01', '2019-02', '2019-03', '2019-04', '2019-05', '2019-06', '2019-07', '2019-08', '2019-09', '2019-10', '2019-11', '2019-12']]

col = df_zill_rent.loc[: , '2019-01':'2019-12']

df_zill_rent['rent_index'] = col.mean(axis=1)

df_zill_rent_less = df_zill_rent[['RegionID', 'rent_index']]
df_zill_rent_less

df_zill_rent_less = df_zill_rent_less.join(df_zill_j.set_index('RegionID'), on='RegionID')[['StateName', 'rent_index']]
df_zill_rent_less.head()
```

```python
df_2018_zill = df_2018_zill.join(df_zill_rent_less.set_index('StateName'), on='state')
df_2018_zill.head()
```

```python
# create new csv of processed data
# this creates a csv that is ~150 MB - too big for Git
df_2018_zill.to_csv('2018_new-var_zillow.csv',index=False)
```
