# Joining the 2019 data (with new variables) to health insurance coverage rates


```python
import pandas as pd
import numpy as np
```

```python
!pwd
```

```python
df_2019 = pd.read_csv("../final_data/2019_new-var-PUMA_mapped_processed.csv")
df_2019.head()
```

```python
df_hv = pd.read_csv("health_insurance_coverage.csv").drop(columns=['Geographic Area Name']).rename(columns={"Estimate-Percent Insured": "insured_rate", "Estimate-Percent Uninsured": "uninsured_rate"})
df_hv.head()
```

```python
# make a copy of df_2018 that we will join to

df_2019_hv = df_2019.copy()
df_2019_hv.head()
```

```python
# try a join

df_2019_hv = df_2019_hv.join(df_hv.set_index('PUMA'), on='full PUMA')
df_2019_hv.head()
```

```python
# create new csv of processed data
# this creates a csv that is ~150 MB - too big for Git
df_2019_hv.to_csv('../final_data/2019_new-var_health-insurance-coverage.csv',index=False)
```
