```python
df_2019 = pd.read_csv("/Users/alexandrawest/Desktop/W210-Capstone/new-var-2019/2019_new-var-PUMA_mapped_processed_2.csv")
df_2019.head()
```

```python
df_hic = pd.read_csv("/Users/alexandrawest/Desktop/W210-Capstone/Census_data/health_insurance_coverage_2.csv").drop(columns=['Geographic Area Name']).rename(columns={"Estimate!!Percent Insured!!Civilian noninstitutionalized population": "insured_rate", "Estimate!!Percent Uninsured!!Civilian noninstitutionalized population": "uninsured_rate"})
df_hic.head()
```

```python
# make a copy of df_2019 that we will join to

df_2019_hic = df_2019.copy()
df_2019_hic.head()
```

```python
# try a join

df_2019_hic = df_2019_hic.join(df_hic.set_index('PUMA'), on='full PUMA')
df_2019_hic.head()
```

```python
# check for null values

df_2019_hic['insured_rate'].isnull().values.any()
```

```python
# show rows where 'insured_rate' is null

df_2019_hic_NULL = df_2019_hic[df_2019_hic['insured_rate'].isnull()]
df_2019_hic_NULL.head(30)
```

```python
# create new csv of processed data
# this creates a csv that is ~150 MB - too big for Git
df_2019_hic.to_csv('/Users/alexandrawest/Desktop/W210-Capstone/joins/2019_new-var_health-insurance-coverage.csv',index=False)
```
