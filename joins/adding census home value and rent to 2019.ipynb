# Joining the 2019 data (with new variables) to home values and gross rent by PUMA


```python
import pandas as pd
import numpy as np
```

```python
df_2019 = pd.read_csv("../final_data/2019_new-var-PUMA_mapped_processed.csv")
df_2019.head()
```

```python
df_hv = pd.read_csv("median_home_value.csv").drop(columns=['Geographic Area Name', 'Margin of Error!!Median value (dollars)']).rename(columns={"Estimate!!Median value (dollars)": "med_home_value"})
df_hv.head()
```

```python
# make a copy of df_2018 that we will join to

df_2019_hv = df_2019.copy()
df_2019_hv.head()
```

```python
# try a join

df_2019_hv = df_2019_hv.join(df_hv.set_index('PUMA'), on='full PUMA')
df_2019_hv.head()
```

```python
df_gr = pd.read_csv("median_gross_rent.csv").drop(columns=['Geographic Area Name', 'Margin of Error!!Median gross rent']).rename(columns={"Estimate!!Median gross rent": "med_gross_rent"})
df_gr.head()
```

```python
# try a join

df_2019_hv = df_2019_hv.join(df_gr.set_index('PUMA'), on='full PUMA')
df_2019_hv.head()
```

```python
# create new csv of processed data
# this creates a csv that is ~150 MB - too big for Git
df_2019_hv.to_csv('../final_data/2019_new-var_home-value_gross-rent.csv',index=False)
```
